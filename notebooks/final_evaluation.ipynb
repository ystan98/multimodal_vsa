```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('final_result.csv')
df['image_pred'] = df['predicted_mean_pool'].apply(lambda x: 0 if x == 'Negative' else 1)
del df['predicted_mean_pool']
df['overall_pred'] = df['text_pred'] + df['audio_pred'] + df['image_pred']
df['label'] = df['label'].apply(lambda x: 0 if x == 'Negative' else 1)
df['overall_pred_basic'] = df['overall_pred'].apply(lambda x: 1 if x >2 else 0)
from sklearn.metrics import confusion_matrix, classification_report

# Define label names
labels = ['Negative', 'Positive']

# Compute confusion matrix
conf_matrix = confusion_matrix(df['label'], df['overall_pred_basic'])

# Generate classification report
class_report = classification_report(df['label'], df['overall_pred_basic'], target_names=labels)

# Display the confusion matrix and classification report
print("Confusion Matrix:\n", conf_matrix)
print("\nClassification Report:\n", class_report)
print("all")
```

```python
conditions = [
    (df['overall_pred'] == 3),
    (df['overall_pred'] == 2),
    (df['overall_pred'] == 1),
    (df['overall_pred'] == 0)
]

categories = ['Positive', 'Slight Negative', 'Negative', 'Strong Negative']

# Assign the categories based on the conditions
df['voting_mechanism'] = np.select(conditions, categories, default='Uncategorized')

# Group by 'label' and 'voting_mechanism', then get value counts and percentages
grouped_data = df.groupby('label')['voting_mechanism'].value_counts(normalize=True).mul(100).round(2)

# Display the result
print(grouped_data)
```
